# Visualisation of the Data

# Count of the dataset based on classification

```python
plt.title("Count of the Classification",fontsize =12)
sns.countplot(Iris_data["Classification"])
plt.xlabel("Classification of the Species", fontsize =12)
plt.ylabel("Count of the classification", fontsize =12)
```

# Pairplot of the entire data set based on the classification

```python
sns.pairplot(Iris_data,hue="Classification", markers=["o", "*", "^"])
```

# Box plot of the Data set - Each column parameter vs Classification

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
sns.boxplot(x="Classification",y="SL",data=Iris_data)
plt.subplot(2,2,2)
sns.boxplot(x="Classification",y="SW",data=Iris_data)
plt.subplot(2,2,3)
sns.boxplot(x="Classification",y="PL",data=Iris_data)
plt.subplot(2,2,4)
sns.boxplot(x="Classification",y="PW",data=Iris_data)

plt.tight_layout()
```

# Violin Plot of Dataset -Each column parameter with Classification

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
sns.violinplot(x="Classification",y="SL",data=Iris_data)
plt.subplot(2,2,2)
sns.violinplot(x="Classification",y="SW",data=Iris_data)
plt.subplot(2,2,3)
sns.violinplot(x="Classification",y="PL",data=Iris_data)
plt.subplot(2,2,4)
sns.violinplot(x="Classification",y="PW",data=Iris_data)
plt.tight_layout()
```

```python
plt.figure(figsize=(30,20))
Iris_data.hist(color="red", rwidth=0.9)
plt.tight_layout()
```
